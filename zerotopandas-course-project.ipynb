### lauches by organization

```python
df = data['Company Name'].value_counts().reset_index()
df.columns = ['company', 'number of lauches']
df = df.sort_values(by='number of lauches')
plt.figure(figsize=(14,17))
y=df["number of lauches"]
x=df["company"]
plt.barh(x,y,color ='r',label='lauches by company')
plt.title("lauches by company",fontsize = 15)
plt.xlabel("no.of launches",fontsize = 15)
plt.ylabel("comapny name",fontsize = 15)
plt.show()
```

### number of active project and retired project

```python
df2 = data["Status Rocket"].value_counts().reset_index()
x = df2["Status Rocket"]
plt.pie(x,labels=df2["index"],explode=[0.1,0.1],autopct="%0.2f%%",shadow=True, radius = 1.5)
plt.show()
```

### success of all company launch

```python
df3 = data["Status Mission"].value_counts().reset_index()
x = df3["index"]
y = df3["Status Mission"]
plt.bar(x,y)
plt.show()
```

###  sucees statistic top 5 company 


```python
df4=pd.crosstab(data["Company Name"],data["Status Mission"]).sort_values(by='Success',ascending=False).reset_index().head(5)
plt.figure(figsize=(10,8))
lst = np.arange(len(df4['Company Name']))

plt.bar(lst,df4['Success'],0.2,color='r',label="Success")
plt.bar(lst+0.2,df4['Partial Failure'],0.2,color='b',label="Partial Failure")
plt.bar(lst+0.4,df4['Failure'],0.2,color='g',label="Failure")
plt.bar(lst+0.6,df4['Prelaunch Failure'],0.2,color='y',label="Prelaunch Failure")
plt.legend()
plt.xticks(lst,df4['Company Name'])

plt.title("status",fontsize = 15)
plt.xlabel("comapny name",fontsize = 15)
plt.ylabel("number",fontsize = 15)

plt.show()
```

### location of program having more than 100 project

```python
df5 = data["Location"].value_counts().reset_index()
df5 =df5[df5.Location>100]
plt.figure(figsize=(8,6))
y=df5["Location"]
x=df5["index"]
plt.barh(x,y,color ='y')
plt.title("location",fontsize = 15)
plt.xlabel("no.of project",fontsize = 15)
plt.ylabel("location",fontsize = 15)
plt.show()
```

Q3. what is success rate of launches by NASA

```python
nasa = data[data["Company Name"]=="NASA"]
a =pd.crosstab(nasa["Company Name"],nasa["Status Mission"])
sucess_rate = a["Success"]/(a["Success"]+a["Partial Failure"]+a["Failure"])
```

```python
print("sucess rate of nasa {}.".format(sucess_rate*100))
```
